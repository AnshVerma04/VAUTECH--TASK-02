```python
# Overall completeness of the dataset
total_cells = df.size
missing_cells = df.isnull().sum().sum()

completeness_percentage = ((total_cells - missing_cells) / total_cells) * 100
completeness_percentage
```

```python
# column - wise completeness
(1 - df.isnull().mean()) * 100
```
